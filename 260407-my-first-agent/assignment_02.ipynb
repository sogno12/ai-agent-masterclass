```python
import os
import openai

OLLAMA_API_URL = os.getenv("OLLAMA_API_URL")

# 클라이언트 초기화
client = openai.OpenAI(
    base_url=OLLAMA_API_URL, # OpenAI API 대신 ollama 사용
    api_key="ollama-key" # 필수로 문자열을 요구하므로 주석 해제 유지
)

# Ollama 서버의 모델 목록 조회
models = client.models.list()

# 사용 가능한 모델의 ID(이름) 출력
print("=== 사용 가능한 모델 목록 ===")
for model in models.data:
    print("-", model.id)
```

```python
import openai

client = openai.OpenAI(
    base_url=OLLAMA_API_URL,
    api_key="ollama-key" 
)

messages = [] # 메모리
```

```python
def call_llm(user_input):
    print(f"\n👤 [사용자 입력]: \"{user_input}\"")
    messages.append({"role":"user", "content": user_input})

    response = client.chat.completions.create(
        model="gemma4:e2b",
        messages=messages
    )
    
    llm_message = response.choices[0].message.content
    print(f"🤖 [답변]: {llm_message}")
    messages.append({"role":"assistant", "content": llm_message})
```

```python
user_messages = [
    "나는 SF 영화를 좋아해",
    "인셉션이랑 인터스텔라는 이미 봤어",
    "오늘 밤에 뭐 볼지 추천해 줄래?",
    "내가 좋아하는 장르랑 이미 본 영화가 뭐라고 했지?"
]
```

```python
for message in user_messages:
    call_llm(message)
```
